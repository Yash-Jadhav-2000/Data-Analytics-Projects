```python
# Write a Python function to identify unique values in a list and return them as a new list
def unique_values(n):
  return list(set(n))

print(unique_values([1,2,3,3,3,3,4,5]))
```

```python
#Write a Python function to reverse a string.
def reverse_string(n):
  return n[::-1]

print(reverse_string("car"))
```

```python
# Given an array of integers, return indices of the two numbers such that they add up to a specific target.
def two_sum(nums, target):
    lookup = {}
    for i, num in enumerate(nums):
        complement = target - num
        if complement in lookup:
            return [lookup[complement], i]
        lookup[num] = i
    return None

print(two_sum([2, 7, 11, 15], 9))
```

```python
#6. Reverse Digits of an Integer
def reversed_int(n):
  sign = -1 if n < 0 else 1
  reverse_n = int(str(abs(n))[::-1])
  return sign * reverse_n

print(reversed_int(-123))
```

```python
# Check if Palindrome
def pali_check(n):
  clean_n = n.lower().replace(" ","")
  if clean_n == n[::-1]:
    print("It is a palindrome")
  else:
    print("Not a palindrom")

pali_check("racecars")
```

```python
def is_palindrome(s):
    clean_s = s.lower().replace(" ", "")
    return clean_s == clean_s[::-1]
print(is_palindrome("race car"))
```

```python
# remove duplicates
def remove_duplicates_ordered(lst):
    seen = set()
    return [x for x in lst if not (x in seen or seen.add(x))]

print(remove_duplicates_ordered([1,2,2,3,4,4,5]))
```

```python
# fibonnaci
def fibonacci_nth(n):
    a, b = 0, 1
    for _ in range(n):
        print(a)
        a, b = b, a + b
    return a+ b

fibonacci_nth(10)
```
